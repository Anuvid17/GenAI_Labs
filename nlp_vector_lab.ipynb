```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv("Reviews.csv", nrows=50000)

print("Rows Loaded:", len(df))
```

Filter Reviews Containing Light

```python
light_reviews = df[
    df["Text"].str.contains(r"\blight\b", case=False, na=False)
].copy()

print("Reviews containing 'light':", len(light_reviews))
```

Display Candidate Reviews

```python
for i, review in enumerate(light_reviews["Text"].head(100)):
    print(f"\nReview {i+1}")
    print(review)
    print("-" * 100)
```

```python
sentence_1 = light_reviews.iloc[1]["Text"]

sentence_2 = light_reviews.iloc[67]["Text"]

sentence_3 = light_reviews.iloc[45]["Text"]

sentences = [sentence_1, sentence_2, sentence_3]

print("\nSelected Sentences\n")

for i, sentence in enumerate(sentences, start=1):
    print(f"Sentence {i}:")
    print(sentence)
    print("-" * 80)
```

```python
vectorizer = CountVectorizer()

document_term_matrix = vectorizer.fit_transform(sentences)

dense_matrix = document_term_matrix.toarray()

print("\nVocabulary:")
print(vectorizer.get_feature_names_out())

print("\nDocument-Term Matrix:")
print(dense_matrix)
```

```python
vec_1 = dense_matrix[0]
vec_2 = dense_matrix[1]
vec_3 = dense_matrix[2]
```

```python
similarity_1_2 = np.dot(vec_1, vec_2)

similarity_1_3 = np.dot(vec_1, vec_3)
```

```python
print("\nSimilarity Scores")
print("=" * 50)

print("Sentence 1 vs Sentence 2 (Different Contexts):",
      similarity_1_2)

print("Sentence 1 vs Sentence 3 (Similar Contexts):",
      similarity_1_3)
```

```python
vocab = vectorizer.get_feature_names_out()

common_12 = [
    vocab[i]
    for i in range(len(vocab))
    if vec_1[i] > 0 and vec_2[i] > 0
]

common_13 = [
    vocab[i]
    for i in range(len(vocab))
    if vec_1[i] > 0 and vec_3[i] > 0
]

print("\nCommon Words (Sentence 1 & Sentence 2):")
print(common_12)

print("\nCommon Words (Sentence 1 & Sentence 3):")
print(common_13)
```
